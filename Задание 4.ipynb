```python
import numpy as np
import pandas
import matplotlib.pyplot as mpt
x = np.array(range(-10,11))
y = 1/x
oxy = mpt.gca()
mpt.plot(x, y)
oxy.spines['bottom'].set_position(('data',0))
oxy.spines['left'].set_position(('data',0))
mpt.title('Гипербола')
mpt.xlabel('<Ось X>')
oxy.xaxis.set_label_coords(0.95, 0.4)
mpt.ylabel('<Ось Y>') 
oxy.yaxis.set_label_coords(0.4, 0.89)
mpt.grid(True)
mpt.xlim(-10,10)
mpt.show()
```
